```python
trestbpsdata = data['trestbps']
type(trestbpsdata)
print("Mean TrestBPS")
statistics.mean(trestbpsdata)
trestbpsdata = pd.to_numeric(trestbpsdata, errors='coerce')
trestbpsdata.fillna(trestbpsdata.mean(), inplace=True)
trestbpsdata.dropna(inplace=True)
mean_trestbps = trestbpsdata.mean()
print(mean_trestbps)
```

```python
print("Median Trestbps")
statistics.median(trestbpsdata)
```

```python
print("Mode Trestbps")
statistics.mode(trestbpsdata)
```

```python
print("Variance Trestbps")
statistics.variance(trestbpsdata)
trestbpsdata = pd.to_numeric(trestbpsdata, errors='coerce')
trestbpsdata.fillna(trestbpsdata.var(), inplace=True)
trestbpsdata.dropna(inplace=True)
mean_trestbps = trestbpsdata.var()
print(mean_trestbps)
```

```python
print("Devisiasi standar uTrestbps")
statistics.stdev(trestbpsdata)
```

```python
print("Skewness Trestbps")
stats.skew(trestbpsdata)
```

```python
print("Kurtosis Trestbps")
stats.kurtosis(trestbpsdata)
```
